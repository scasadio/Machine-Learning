```python
import pandas as pd
from sklearn import naive_bayes, metrics
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

sms_data= pd.read_csv('datasets/smsspamcollection/SMSSpamCollection', 
                      header=None,
                      sep='\t',
                      names=['Label','SMS']) #diamo noi il nome alle colonne perche non ce li ha 

print(sms_data.groupby('Label').count())

for i in range(0,5):
    print(sms_data['Label'][i])
    print(sms_data['SMS'][i])
    print("\n")

sms_data_clean=sms_data.copy()
sms_data_clean['SMS'] = sms_data_clean['SMS'].str.replace(r'\W+', ' ', regex=True).str.replace(r'\s+', ' ',  regex=True).str.strip()
sms_data_clean['SMS'] = sms_data_clean['SMS'].str.lower()
sms_data_clean['SMS'] = sms_data_clean['SMS'].str.split()

for i in range(0,5):
    print(sms_data_clean['Label'][i])
    print(sms_data_clean['SMS'][i])
    print("\n")
```

.str.replace(r'\W+', ' ', regex=True) -> sostituisce tutti i caratteri non-alfanumerici (simboli, punteggiatura, ecc.) con uno spazio vuoto (' '). L'espressione regolare \W+ identifica tutti i caratteri non-alfanumerici. Il parametro regex=True specifica che si sta utilizzando un'espressione regolare per trovare e sostituire i pattern.

.str.replace(r'\W+', ' ', regex=True) -> sostituisce tutti gli spazi multipli con uno spazio singolo. L'espressione regolare \s+ trova tutti i gruppi di spazi consecutivi e li sostituisce con uno spazio singolo.

.str.strip() -> rimuove eventuali spazi vuoti in eccesso all'inizio e alla fine di ciascuna stringa.

.str.lower() -> converte tutti i caratteri all'interno di ogni stringa nella colonna 'SMS' in caratteri minuscoli. Ciò uniforma la formattazione dei testi, rendendo più facile l'analisi e la ricerca di pattern.

.str.split() -> suddivide ogni stringa nella colonna 'SMS' in una lista di parole. Ogni parola diventa un elemento della lista. Questo può essere utile per analizzare il testo parola per parola, applicare analisi lessicali o creare modelli che trattano le parole come unità di base.

```python
le=preprocessing.LabelEncoder()
sms_data_clean['Label']=le.fit_transform(sms_data_clean['Label'])
#assegno valori 0 e 1 a ham e spam
for i in range(0,5):
    print(sms_data_clean['Label'][i])
    print(sms_data_clean['SMS'][i])
    print("\n")
```

```python
X_train, X_test, y_train, y_test = train_test_split(sms_data_clean['SMS'], 
                                                    sms_data_clean['Label'], 
                                                    test_size=0.2,
                                                    random_state=42)

sms_data_clean['Label'].value_counts()
```

```python
sms_data_clean['Label'].value_counts() /sms_data.shape[0] *100 #/numero rigghe*100
```

```python
print(X_train.sum())

#list comprehension 

vocabulary = list(set(X_train.sum())) #list(set) lista che contiene elementi in maniera univoca
    #Questa riga crea un vocabolario unico estratto dai dati di addestramento X_train. X_train 
X_train_voc = pd.DataFrame([
    [row.count(word) for word in vocabulary]
    for row in X_train], columns=vocabulary)
    # crea un DataFrame X_train_voc che rappresenta la frequenza di ogni parola nel vocabolario in ciascun documento di addestramento. 
    #Ogni riga del DataFrame corrisponde a un documento di addestramento e ogni colonna corrisponde a una parola nel vocabolario. 
    #Il valore in ogni cella del DataFrame è la frequenza della parola corrispondente nel documento.
X_test_voc = pd.DataFrame([
    [row.count(word) for word in vocabulary]
    for row in X_test], columns=vocabulary)

X_train_voc.head()
```

```python
#altro modo meno compatto
m=[]
for row in X_train:
    l=[]
    for word in vocabulary:
        l.append(row.count(word))

    m.append(l)

X_train_voc_2=pd.DataFrame(m,columns=vocabulary)
X_train_voc_2.head()
```

```python
print(X_train_voc.shape)
print(X_test_voc.shape)
```

```python
cl_gauss=naive_bayes.GaussianNB()
res_gauss=cl_gauss.fit(X_train_voc,y_train).predict(X_test_voc)
print(metrics.accuracy_score(y_test, res_gauss)*100)
```

Funziona molto bene in caso di distribuzione gaussiana e variabili continue.

```python
from sklearn import naive_bayes
cl_multi=naive_bayes.MultinomialNB()
res_multi=cl_multi.fit(X_train_voc,y_train).predict(X_test_voc)
print(metrics.accuracy_score(y_test, res_multi)*100)
```

Funziona molto bene con componenti discrete quindi è il migliore in questo caso.

```python
cl_bern=naive_bayes.BernoulliNB()
res_bern=cl_bern.fit(X_train_voc,y_train).predict(X_test_voc)
print(metrics.accuracy_score(y_test, res_bern)*100)
```
